```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('purdue.csv')
print(data.head())
print(data[0:5])
```

```python
# Replace the nan values in Loc with "H"
data['Loc'] = data['Loc'].fillna('H')
print(data.head())

average_purdue_fouls = data['PF'].mean()
average_opponent_fouls = data['OPF'].mean()

total_purdue_fouls = data['PF'].sum()
total_opponent_fouls = data['OPF'].sum()

average_purdue_ft_attempts = data['FTA'].mean()
average_opponent_ft_attempts = data['OFTA'].mean()

total_purdue_ft_attempts = data['FTA'].sum()
total_opponent_ft_attempts = data['OFTA'].sum()
```

```python
# Create a 2x2 grid of plots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Plot average Purdue fouls vs. average opponent fouls
axs[0, 0].bar(['Purdue', 'Opponent'], [average_purdue_fouls, average_opponent_fouls])
axs[0, 0].set_title('Average Fouls per Game')

# Plot total Purdue fouls vs. total opponent fouls
axs[0, 1].bar(['Purdue', 'Opponent'], [total_purdue_fouls, total_opponent_fouls])
axs[0, 1].set_title('Total Fouls')

# Plot average Purdue free throw attempts vs. average opponent free throw attempts
axs[1, 0].bar(['Purdue', 'Opponent'], [average_purdue_ft_attempts, average_opponent_ft_attempts])
axs[1, 0].set_title('Average Free Throw Attempts per Game')

# Plot total Purdue free throw attempts vs. total opponent free throw attempts
axs[1, 1].bar(['Purdue', 'Opponent'], [total_purdue_ft_attempts, total_opponent_ft_attempts])
axs[1, 1].set_title('Total Free Throw Attempts')

plt.tight_layout()
plt.show()
```

```python
indiana_data = pd.read_csv('indiana.csv')

# Replace the nan values in Loc with "H"
indiana_data['Loc'] = indiana_data['Loc'].fillna('H')

average_indiana_fouls = indiana_data['PF'].mean()

total_indiana_fouls = indiana_data['PF'].sum()

average_indiana_ft_attempts = indiana_data['FTA'].mean()

total_indiana_ft_attempts = indiana_data['FTA'].sum()
```

```python
# Create a 2x2 grid of plots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
# Plot average Purdue fouls vs. average Indiana fouls. Set the Purdue bar to #CEB888 and the Indiana bar to #990000
axs[0, 0].bar(['Purdue', 'Indiana'], [average_purdue_fouls, average_indiana_fouls], color=['#CEB888', '#990000'])
axs[0, 0].set_title('Average Fouls per Game')

# Plot total Purdue fouls vs. total Indiana fouls
axs[0, 1].bar(['Purdue', 'Indiana'], [total_purdue_fouls, total_indiana_fouls], color=['#CEB888', '#990000'])
axs[0, 1].set_title('Total Fouls')

# Plot average Purdue free throw attempts vs. average Indiana free throw attempts
axs[1, 0].bar(['Purdue', 'Indiana'], [average_purdue_ft_attempts, average_indiana_ft_attempts], color=['#CEB888', '#990000'])
axs[1, 0].set_title('Average Free Throw Attempts per Game')

# Plot total Purdue free throw attempts vs. total Indiana free throw attempts
axs[1, 1].bar(['Purdue', 'Indiana'], [total_purdue_ft_attempts, total_indiana_ft_attempts], color=['#CEB888', '#990000'])
axs[1, 1].set_title('Total Free Throw Attempts')

plt.tight_layout()
plt.show()
```

```python
# Create a point plot with no line of Purdue's fouls per game vs. the opponent's fouls per game
plt.figure(figsize=(10, 6))
plt.plot(data['PF'], label='Purdue', marker='o', color='#CEB888')
plt.plot(data['OPF'], label='Opponent', marker='x', color='#000099')
plt.xlabel('Game Number')
plt.ylabel('Fouls')
plt.title('Purdue Fouls vs. Opponent Fouls')
plt.legend()
plt.show()
```

```python
purdue_foul_diff = data['PF'] - data['OPF']

# Create a bar plot of Purdue foul differential
plt.bar(data['Date'], purdue_foul_diff)
plt.ylabel('Foul Differential')
plt.title('Purdue Foul Differential')
plt.xticks([])
plt.tight_layout()
plt.show()
```

```python
# Create a scatter plot of Purdue free throw attempts vs. Purdue points. Mark the wins with a green dot and the losses with a red dot
colors = []
for i in range(len(data)):
    if 'W' in str(data['W/L'][i]):
        colors.append('green')
    else:
        colors.append('red')
plt.scatter(data['FTA'], data['Tm'], c=colors)
plt.xlabel('Free Throw Attempts')
plt.ylabel('Points')
plt.title('Purdue Free Throw Attempts vs. Points')
plt.tight_layout()
plt.show()
```

```python
purdue_home_foul_avg = data[data['Loc'] == 'H']['PF'].mean()
purdue_away_foul_avg = data[data['Loc'] == '@']['PF'].mean()
purdue_neutral_foul_avg = data[data['Loc'] == 'N']['PF'].mean()

average_home_opponent_fouls = data[data['Loc'] == 'H']['OPF'].mean()
average_away_opponent_fouls = data[data['Loc'] == '@']['OPF'].mean()
average_neutral_opponent_fouls = data[data['Loc'] == 'N']['OPF'].mean()

total_purdue_home_fouls = data[data['Loc'] == 'H']['PF'].sum()
total_purdue_away_fouls = data[data['Loc'] == '@']['PF'].sum()
total_purdue_neutral_fouls = data[data['Loc'] == 'N']['PF'].sum()

total_home_opponent_fouls = data[data['Loc'] == 'H']['OPF'].sum()
total_away_opponent_fouls = data[data['Loc'] == '@']['OPF'].sum()
total_neutral_opponent_fouls = data[data['Loc'] == 'N']['OPF'].sum()

purdue_home_ft_avg = data[data['Loc'] == 'H']['FTA'].mean()
purdue_away_ft_avg = data[data['Loc'] == '@']['FTA'].mean()
purdue_neutral_ft_avg = data[data['Loc'] == 'N']['FTA'].mean()

average_home_opponent_ft_attempts = data[data['Loc'] == 'H']['OFTA'].mean()
average_away_opponent_ft_attempts = data[data['Loc'] == '@']['OFTA'].mean()
average_neutral_opponent_ft_attempts = data[data['Loc'] == 'N']['OFTA'].mean()

total_purdue_home_ft_attempts = data[data['Loc'] == 'H']['FTA'].sum()
total_purdue_away_ft_attempts = data[data['Loc'] == '@']['FTA'].sum()
total_purdue_neutral_ft_attempts = data[data['Loc'] == 'N']['FTA'].sum()

total_home_opponent_ft_attempts = data[data['Loc'] == 'H']['OFTA'].sum()
total_away_opponent_ft_attempts = data[data['Loc'] == '@']['OFTA'].sum()
total_neutral_opponent_ft_attempts = data[data['Loc'] == 'N']['OFTA'].sum()

# Create a 2x2 grid of plots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

barWidth = 0.25
r1 = np.arange(3)
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
axs[0, 0].bar(r1, [purdue_home_foul_avg, purdue_away_foul_avg, purdue_neutral_foul_avg], color='#CEB888', width=barWidth, label='Purdue')
axs[0, 0].bar(r2, [average_home_opponent_fouls, average_away_opponent_fouls, average_neutral_opponent_fouls], color='#000099', width=barWidth, label='Opponent')
axs[0, 0].set_title('Average Fouls per Game by Location')
axs[0, 0].set_xticks([r + barWidth for r in range(3)])
axs[0, 0].set_xticklabels(['Home', 'Away', 'Neutral'])
axs[0, 0].bar_label(axs[0, 0].containers[0])
axs[0, 0].bar_label(axs[0, 0].containers[1])
axs[0, 0].legend(ncols=2)

axs[0, 1].bar(r1, [total_purdue_home_fouls, total_purdue_away_fouls, total_purdue_neutral_fouls], color='#CEB888', width=barWidth, label='Purdue')
axs[0, 1].bar(r2, [total_home_opponent_fouls, total_away_opponent_fouls, total_neutral_opponent_fouls], color='#000099', width=barWidth, label='Opponent')
axs[0, 1].set_title('Total Fouls by Location')
axs[0, 1].set_xticks([r + barWidth for r in range(3)])
axs[0, 1].set_xticklabels(['Home', 'Away', 'Neutral'])
axs[0, 1].bar_label(axs[0, 1].containers[0])
axs[0, 1].bar_label(axs[0, 1].containers[1])
axs[0, 1].legend(ncols=2)

axs[1, 0].bar(r1, [purdue_home_ft_avg, purdue_away_ft_avg, purdue_neutral_ft_avg], color='#CEB888', width=barWidth, label='Purdue')
axs[1, 0].bar(r2, [average_home_opponent_ft_attempts, average_away_opponent_ft_attempts, average_neutral_opponent_ft_attempts], color='#000099', width=barWidth, label='Opponent')
axs[1, 0].set_title('Average Free Throw Attempts per Game by Location')
axs[1, 0].set_xticks([r + barWidth for r in range(3)])
axs[1, 0].set_xticklabels(['Home', 'Away', 'Neutral'])
axs[1, 0].bar_label(axs[1, 0].containers[0])
axs[1, 0].bar_label(axs[1, 0].containers[1])
axs[1, 0].legend(ncols=2)

axs[1, 1].bar(r1, [total_purdue_home_ft_attempts, total_purdue_away_ft_attempts, total_purdue_neutral_ft_attempts], color='#CEB888', width=barWidth, label='Purdue')
axs[1, 1].bar(r2, [total_home_opponent_ft_attempts, total_away_opponent_ft_attempts, total_neutral_opponent_ft_attempts], color='#000099', width=barWidth, label='Opponent')
axs[1, 1].set_title('Total Free Throw Attempts by Location')
axs[1, 1].set_xticks([r + barWidth for r in range(3)])
axs[1, 1].set_xticklabels(['Home', 'Away', 'Neutral'])
axs[1, 1].bar_label(axs[1, 1].containers[0])
axs[1, 1].bar_label(axs[1, 1].containers[1])
axs[1, 1].legend(ncols=2)

plt.tight_layout()
plt.show()

```
